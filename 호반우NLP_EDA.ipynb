```python
from transformers import AutoTokenizer

max_input_length = 200
#tokenizer = AutoTokenizer.from_pretrained("monologg/koelectra-base-discriminator")
tokenizer = AutoTokenizer.from_pretrained("klue/roberta-large")
```

```python

inputs = train.loc[54]['sentence']
print(inputs)
model_inputs = tokenizer.tokenize(inputs, max_length=28, truncation=True,padding='max_length')


print(model_inputs)
```

```python
inputs = train.loc[54]['sentence']
print(inputs)
model_inputs = tokenizer.encode(inputs, max_length=28, truncation=True,padding='max_length')


print(model_inputs)
```

```python

inputs = train.loc[54]['context']
print(inputs)
model_inputs = tokenizer.tokenize(inputs, max_length=28, truncation=True,padding='max_length')


print(model_inputs)
```

### 토크나이저 관련
- [mecab](https://hong-yp-ml-records.tistory.com/91)

```python
from konlpy.tag import Okt, Mecab

okt = Okt()
mecab = Mecab()
```

```python
okt.morphs('한글 형태소 분석기(오케이티)로 테스트를 해보았습니다. 정상 설치 및 동작이 잘 됩니다.')
```

```python
mecab.morphs('한글 형태소 분석기(미켑)로 테스트를 해보았습니다. 정상 설치 및 동작이 잘 됩니다.')
```

```python

inputs = train.loc[310]['context']
print(inputs)
model_inputs = tokenizer.tokenize(inputs, max_length=28, truncation=True,padding='max_length')
model_inputs2 = mecab.morphs(inputs)
model_inputs3 = okt.morphs(inputs)
model_inputs4 = mecab.nouns(inputs)

model_inputs5 = mecab.pos(inputs)

print(model_inputs)

print(model_inputs2)

print(model_inputs3)

print(model_inputs4)

print(model_inputs5)
```

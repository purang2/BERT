```python
from datasets import load_dataset
import pandas as pd 

data_path ="drive/MyDrive/호반우NLP_한국인감정/KERC22Dataset/KERC22Dataset_PublicTest"


train_data = pd.read_csv(data_path+"/train_data.tsv", sep='\t')
train_labels = pd.read_csv(data_path+"/train_labels.csv")

train_data
```

```python
train_drop = train_data[['context','sentence','person']]

train_pair = train_drop

train_pair['label'] = train_labels['label']

train_pair
```

```python
import numpy as np 

train_pair['context2'] = train_pair['context'].fillna('none')
train_pair['sentence2'] = train_pair['sentence'].fillna('none')
#모든 글자 합치기
train_pair['new'] = train_pair['person'] +" " + train_pair['context2'] + " " + train_pair['sentence2']
#train_pair['new'] = train_pair['sentence'] +" " + train_pair['context'] + " " + train_pair['person']

train_pairs = train_pair[['label','new']]

train_pairs.head()

labels = {'euphoria':0,
          'neutral':1,
          'dysphoria':2
          }

train_pairs['label'] = [labels[label] for label in train_pairs['label']]
```

```python
eval_data = train_df

answer_set = [] #answer set 

labels = ['신체적, 정서적으로 행복한, 유쾌한, 의기양양한, 재미있는, 편안한','잔잔한, 차분한, 보통인, 안정적인','불행한, 불만족한, 안절부절하지 못한, 낙담한, 우울한, 좌절한, 불안한']


prompt = '''
'''


#k-shot 
for i in range(8):
  prompt += f"맥락: {eval_data.iloc[i]['new']}\n 감정: {labels[eval_data.iloc[i]['label']]} [끝]\n\n"





eval_data
```

```python

import gc
import torch 

answer_set = []

for i in range(30,35):
  
#  gc.collect()
#  torch.cuda.empty_cache()
  index = i
  prompt_with_query = prompt + f"맥락: {eval_data.iloc[i]['new']}\n 감정: "



  with torch.no_grad():
    tokens = tokenizer.encode(prompt_with_query, return_tensors='pt')
    gen_tokens = model.generate(tokens, do_sample=True, temperature=0.8, max_length=len(tokens[0])+30)
    generated = tokenizer.batch_decode(gen_tokens)[0]

  #print(generated[len(prompt):])
  answer_set.append(generated[len(prompt_with_query):])
  print(i)
```
